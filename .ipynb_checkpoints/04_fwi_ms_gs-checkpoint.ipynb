```python
criteria = torch.nn.L1Loss(reduction='sum')
criteria1 = torch.nn.L1Loss(reduction='sum')
learn1 = 30#30 #20

vp0 = vp_initial.to(DEVICE)
vp0 = vp0.requires_grad_(True)


optimer = torch.optim.Adam([{'params': [vp0], 'lr': learn1,'betas': (0.9, 0.98)}])
#optimer = AdaHessian([{'params': [vp0],'lr': learn1, 'betas': (0.9, 0.98),'eps': 1e-4,'weight_decay':0.0,'hessian_power':1.0}])
#
#scheduler_transfomerdecoder = torch.optim.lr_scheduler.CyclicLR(optim_transfomerdecoder, base_lr=1.0e-3, max_lr=0.01)

all_loss_data = []
#all_loss_vx_model = []
#all_loss_vy_model = []
#all_loss_rho_model = []
all_loss_model =[]
time_each_iter = []
tv_loss = []
SNR = []
SSIM = []
ERROR = []
```

```python

cpu_mem_log = [] 
gpu_mem_log = []
```

```python

import time
t_start = time.time()

grad_accum = None
num_grads = 0


sigmas = [10,8,6,5,3,0.01]

for simga in sigmas: 
    print('Now inversion on sigma :',simga)
    for iter in range(ITERATION):
        loss_data_minibatch = []
        time_each_bath_start = time.time()
        if (iter + 1) > 1000:
            print(f"Iteration {iter+1}: Update source...")
            src = estimate_source_wavelet(vp_initial = vp0.detach().clone(), 
                                      obs_data  = obs_used, 
                                      src_loc = src_loc, 
                                      rec_loc = rec_loc, 
                                      inpa = inpa, 
                                      Physics = Physics_deepwave, 
                                      DEVICE = DEVICE)
        
            src = src.unsqueeze(1).to(DEVICE)
        else:
            src = src
        
        for batch in range(mini_batches):
        
            optimer.zero_grad()
            src_loc_batch = src_loc[batch::mini_batches]
            rec_loc_batch = rec_loc[batch::mini_batches]
            src_batch = src[batch::mini_batches]
                        
            physics = Physics_deepwave(inpa['dh'], inpa['dt'], src=src_batch,
                                src_loc=src_loc_batch, rec_loc=rec_loc_batch,
                                       accuracy = inpa['accuracy'],
                                    pml_width =  inpa['pml_width']
                                )
                
            with torch.no_grad():
                if submarine == "yes":
                    vp0[:submarine_deep,:] = submarine_vp
                else:
                    vp0 = vp0
                    
                vp0[vp0>vp_true.max()] = vp_true.max()
                vp0[vp0<vp_true.min()] = vp_true.min() 
                
            vp0 = vp0.requires_grad_(True)
            taux_est = physics(vp0.squeeze()) ###[1,shots_batch,nt,receviers]
    
            d_obs1 = d_obs.unsqueeze(0)
      
            taux_est_filtered = taux_est
            d_obs_filtered = d_obs1[:, batch::mini_batches]
            
            lam_tv = 0#0.001
            tvloss = tv_l1_regularization(vp0)
            #taux_est_filtered = (taux_est_filtered - torch.mean(taux_est_filtered)) / torch.std(taux_est_filtered)
            #d_obs_filtered = (d_obs_filtered - torch.mean(d_obs_filtered)) / torch.std(d_obs_filtered)
            kernel_size = int(simga*6+1)|1
            channels = taux_est_filtered.shape[1]
            dtype = torch.float
            kernel = gaussian_kernel(size=kernel_size, channels=channels, sigma=simga, dtype=dtype, device=DEVICE)

            pyramids_obs = gaussian_conv2d(d_obs_filtered,kernel)
            pyramids_syn = gaussian_conv2d(taux_est_filtered,kernel)
            
            if loss_fn == 'l1':
                criteria = torch.nn.L1Loss(reduction='mean')
                weight = 1.0e2
            else:
                criteria =  torch.nn.MSELoss(reduction='mean')
                weight = 1.0e4

            loss_data = weight*criteria(pyramids_obs,pyramids_syn)
    
            loss = loss_data
            loss.backward()
            
            optimer.step()
                  
            loss_data = loss.item()
        
        all_loss_data.append(np.mean(loss_data))
        time_each_bath_end = time.time()
        tv_loss.append(lam_tv*tvloss)
        
        time_each_iter.append(time_each_bath_end - time_each_bath_start)
        
        mem_now_cpu = get_cpu_memory()
        cpu_mem_log.append(mem_now_cpu)
        mem_now_gpu = get_gpu_memory()
        gpu_mem_log.append(mem_now_gpu)
        
        snr = ComputeSNR(vp0.detach().cpu().numpy(), \
                      vp_true.detach().cpu().numpy())
        SNR = np.append(SNR, snr)
     
        ssim = ComputeSSIM(vp0.detach().cpu().numpy(), \
                      vp_true.detach().cpu().numpy())
        SSIM = np.append(SSIM, ssim)
    
        
        
        with torch.no_grad():
            all_loss_model.append(
                criteria1(vp0, vp_true).item()
            )
        if (iter+1)%1 ==0:
            print(f"Iter {iter + 1} = loss: {all_loss_data[-1]:.4f},model: {all_loss_model[-1]:.4f},time_iter:{time_each_iter[-1]:.4f},tvloss:{tv_loss[-1]:.2f},snr_vp:{SNR[-1]:.3f}")
        if (iter+1)%10 ==0:
            np.save(vp_save_path + 'recx_iter_%s_sigma_%s.npy' % (iter + 1, simga), vp0.cpu().detach().numpy())
           # np.save(src_path + 'recx_iter_%s_s%s.npy' % (iter+1, level), src.cpu().detach().numpy())
        if (iter+1)%10 == 0:
            
            #np.save('./reconstructions/segx156x384_fwi/fwi_adahessian_iter_%s.npy' %(iter+1), m.cpu().detach().numpy(), 2)
            
            fig, ax = plt.subplots(1,1, figsize=(4, 4))
            im = ax.imshow(vp0.detach().cpu(), 
                              cmap='jet', vmin=vp0.min(), 
                       vmax=vp0.max()) 
            points = ax.get_position().get_points()
            dy = points[1, 1] - points[0, 1]
    
            cax = fig.add_axes([0.91, points[0, 1], 0.02, dy])
            cax.yaxis.set_ticks_position("right")
            cbar = fig.colorbar(im, cax=cax, orientation="vertical", 
                                extend="neither", label="$V_P (m/s)$"
                                )
            plt.show()
        #scheduler_transfomerdecoder.step()    
t_end = time.time()
elapsed_time = t_end - t_start
print('Running complete in {:.0f}m  {:.0f}s' .format(elapsed_time //60 , elapsed_time % 60))
```
